```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v = np.linspace(3,4.2,13)*0.51444 # vitesse en m/s
Tc = np.array([5429, 5692, 5954, 6223, 6513, 6861, 7110, 7408, 7749, 8076, 8410, 8760, 9103])*9.81 # Traction en N
```

Affichage : Tc = f(v)

```python
plt.scatter(v, Tc)
plt.xlabel("v")
plt.ylabel("Tc")
plt.title("Tc = f(v)")
plt.show()
```

Régression linéaire

```python
coef = np.polyfit(v, Tc, 1)          # coef[0]=pente, coef[1]=ordonnée à l'origine
a, b = coef
print(f"Tc ≈ {a:.4f}·v + {b:.4f}")
```

Affichage de la droite

```python
Tc_fit = a*v + b
plt.scatter(v, Tc)
plt.plot(v, Tc_fit, 'r')
plt.show()
```

Régression quadratique

```python
coef = np.polyfit(v, Tc, 2)
a, b, c = coef
print(f"Tc ≈ {a:.4f}·v² + {b:.4f}·v + {c:.4f}")
```

Affichage de la droite

```python
Tc_fit = a*v**2 + b*v + c
plt.scatter(v, Tc)
plt.plot(v, Tc_fit, 'r')
plt.show()
```
